# combine all symptoms

```python
symptom_columns=["Symptom 1","Symptom 2","Symptom 3","Symptom 4","Symptom 5"]
df["All_symptoms"]=df[symptom_columns].values.tolist()
df.drop(columns=symptom_columns)
```

```python
df.columns
```

# MultiLabelBinarizer

```python
from sklearn.preprocessing import MultiLabelBinarizer
mlb=MultiLabelBinarizer()
symptom_encoded=mlb.fit_transform(df["All_symptoms"])
drop_cols=symptom_columns.copy()
drop_cols.append("All_symptoms")
```

```python
drop_cols
```

```python
df.drop(columns=drop_cols,inplace=True)
```

```python
df.head(3)
symptom_encoded.dtype
```

```python
symptom_encoded_df=pd.DataFrame(symptom_encoded,columns=mlb.classes_,index=df.index)
```

```python
df=pd.concat([df,symptom_encoded_df],axis=1)
```

```python
df.head()
```

```python
X=df.drop(columns=["Disease"])
Y=df["Disease"]
```

```python
X
```

```python
Y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct=ColumnTransformer([("encoder",OneHotEncoder(),["Dog Breed"])],remainder="passthrough")
X=pd.DataFrame(ct.fit_transform(X))
```

```python
X
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
Y=le.fit_transform(Y)
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=11)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
df["Disease"].value_counts()
```

#  combined all algorithm in one

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,f1_score

models=[DecisionTreeClassifier,KNeighborsClassifier,GaussianNB,RandomForestClassifier]

for model in models:
    md=model()
    md.fit(X_train,Y_train)
    y_pred=md.predict(X_test)
    print("=============",model,"============")
    print("accuracy_score",accuracy_score(Y_test,y_pred)*100)
```

```python
from sklearn.ensemble import AdaBoostClassifier,GradientBoostingClassifier
models=[AdaBoostClassifier,GradientBoostingClassifier]
for model in models:
    md=model()
    md.fit(X_train,Y_train)
    y_pred=md.predict(X_test)
    print("=============",model,"============")
    print("accuracy_score",accuracy_score(Y_test,y_pred)*100)
```
